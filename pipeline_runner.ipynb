# Run Classification Model

```python
import sys
import matplotlib.pyplot as plt
import numpy as np

sys.path.append("./src")
from classification import main as classification_main

# Define hyperparameter lists
horizons = [1, 3, 5]
learning_rates = [0.005, 0.001, 0.0001]
batch_sizes = [32, 64, 128]

# Run everything in classification.py once
results = classification_main(
    horizon=horizons,
    learning_rate=learning_rates,
    batch_size=batch_sizes
)

# Unpack return
results_by_model = results["results_by_model"]
dnn_training_stats = results["dnn_training_stats"]  # dict[h] = {losses, train_accs, test_accs, config}

# Plot metrics by model and horizon
for metric in ["Precision", "Recall", "Accuracy", "F1-Score"]:
    plt.figure(figsize=(8, 5))
    for model_name in results_by_model:
        y = [results_by_model[model_name][h][metric] for h in horizons]
        plt.plot(horizons, y, label=model_name, marker='o')
    plt.title(f"{metric} by Model Across Horizons")
    plt.xlabel("Horizon")
    plt.ylabel(metric)
    plt.xticks(horizons)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Plot DNN learning curves (loss & acc) for each horizon
for h in horizons:
    stats = dnn_training_stats[h]
    label = f"DNN - Horizon {h} (lr={stats['config']['lr']}, bs={stats['config']['batch_size']})"
    epochs = range(1, len(stats["train_losses"]) + 1)

    fig, ax = plt.subplots(1, 2, figsize=(12, 4))

    # Training loss
    ax[0].plot(epochs, stats["train_losses"], label="Train Loss")
    ax[0].set_title(f"Training Loss - {label}")
    ax[0].set_xlabel("Epoch")
    ax[0].set_ylabel("Loss")
    ax[0].grid(True)

    # Accuracy
    ax[1].plot(epochs, stats["train_accuracies"], label="Train Acc")
    ax[1].plot(epochs, stats["test_accuracies"], label="Test Acc")
    ax[1].set_title(f"Accuracy - {label}")
    ax[1].set_xlabel("Epoch")
    ax[1].set_ylabel("Accuracy")
    ax[1].legend()
    ax[1].grid(True)

    plt.tight_layout()
    plt.show()
```

# Run Regression Model

```python
from regression import main as regression_main

horizons = [1, 3, 5]
learning_rates = [0.005, 0.001, 0.0001]
batch_sizes = [32, 64, 128]

results = regression_main(horizons=horizons, learning_rates=learning_rates, batch_sizes=batch_sizes)

results_by_model = results["results_by_model"]
dnn_training_stats = results["dnn_training_stats"]

metrics = ["MAE", "MSE", "R2"]
models = ["LinearRegression", "RandomForest", "PyTorchDNN"]

#1. Line plots across horizons
for metric in metrics:
    plt.figure(figsize=(8, 5))
    for model in models:
        values = [results_by_model[h][model][metric] for h in horizons]
        plt.plot(horizons, values, marker="o", label=model)
    plt.title(f"{metric} by Model Across Horizons")
    plt.xlabel("Horizon (days)")
    plt.ylabel(metric)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# 2. Bar plots across horizons
for metric in metrics:
    plt.figure(figsize=(8, 5))
    width = 0.25
    x = np.arange(len(horizons))
    for i, model in enumerate(models):
        values = [results_by_model[h][model][metric] for h in horizons]
        plt.bar(x + i * width, values, width=width, label=model)
    plt.xticks(x + width, [f"H{h}" for h in horizons])
    plt.title(f"{metric} by Model")
    plt.ylabel(metric)
    plt.xlabel("Horizon")
    plt.legend()
    plt.tight_layout()
    plt.show()

# 3. DNN learning curves
for h in horizons:
    stats = dnn_training_stats[h]
    label = f"H{h} (lr={stats['config']['lr']}, bs={stats['config']['batch_size']})"
    epochs = range(1, len(stats["train_losses"]) + 1)

    fig, ax = plt.subplots(1, 2, figsize=(12, 4))

    ax[0].plot(epochs, stats["train_losses"], label="Train Loss")
    ax[0].set_title(f"Training Loss - {label}")
    ax[0].set_xlabel("Epoch")
    ax[0].set_ylabel("Loss")
    ax[0].grid(True)

    ax[1].plot(epochs, stats["train_r2s"], label="Train R²")
    ax[1].plot(epochs, stats["test_r2s"], label="Test R²")
    ax[1].set_title(f"R² Score - {label}")
    ax[1].set_xlabel("Epoch")
    ax[1].set_ylabel("R²")
    ax[1].legend()
    ax[1].grid(True)

    plt.tight_layout()
    plt.show()
```
